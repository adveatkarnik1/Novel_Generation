```python
#chat = ChatGroq(temperature=0, groq_api_key=groq_api_key, model_name="llama2-70b-4096", max_tokens = 1000)
chat = HuggingFaceHub(repo_id ='NousResearch/Nous-Hermes-2-Mixtral-8x7B-DPO', model_kwargs={"max_new_tokens": 600})
print(chat.invoke("Who is Usain Bolt"))
```

```python
embeddings_model_name = "mixedbread-ai/mxbai-embed-large-v1"

embeddings = HuggingFaceEmbeddings(model_name=embeddings_model_name)
```

```python
persist_directory = "/content/drive/MyDrive/chroma_percy_jackson_index"
saved_vectordb = Chroma(persist_directory=persist_directory, embedding_function=embeddings)

print(saved_vectordb)
```

```python
retriever = saved_vectordb.as_retriever(search_type = 'mmr')
# retriever.search_kwargs['fetch_k'] = 20
retriever.search_kwargs['k'] = 20
```

```python
story_template = '''You are a creative and helpful storyteller AI. When given context and question, you get a creative and in depth story based on the question.

context = {context}

question = Write a detailed story based on the context for {question}

Answer:
'''

story_prompt = PromptTemplate(input_variables = ['context','question'], template = story_template)
```

```python
chain_story = RetrievalQA.from_chain_type(
    llm=chat,
    retriever=retriever,
    chain_type="stuff",
    chain_type_kwargs={"prompt": story_prompt},
    verbose=True
)

handler = StdOutCallbackHandler()

print(chain_story.invoke(
    'Who is Percy Jackson?',
    callbacks=[handler]
))
```

```python
print(chain_story.run(
    "Percy going on a date with Annabeth",
    callbacks=[handler]
))
```

```python
question_template = '''You are a smart AI who responds based on the provided context. When given context and question, you get a creative and in depth story based on the question.

context = {context}

question = Write a detailed explanation based on the context for {question}

Answer:
'''

question_prompt = PromptTemplate(input_variables = ['context','question'], template = question_template)
```

```python
chain_query = RetrievalQA.from_chain_type(
    llm=chat,
    retriever=retriever,
    chain_type="stuff",
    chain_type_kwargs={"prompt": question_prompt},
    verbose=True
)

handler = StdOutCallbackHandler()

print(chain_query.invoke(
    'Who is Percy Jackson?',
    callbacks=[handler]
))
```

```python
print(chain_query.run("WHo is Tyson"))
```
